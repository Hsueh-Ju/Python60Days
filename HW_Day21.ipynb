# 範例重點

如何使用亂數, 資料集來操作


```python
import matplotlib.pyplot as plt
import numpy as np


x = np.arange(0,360)
y = np.cos(x*np.pi/180.0)
plt.plot(x,y)
plt.savefig('filename.png',dpi=300,format='png')
plt.show()
```

# 散點圖: Scatter Plots
    
    顏色由（X，Y）的角度給出。

    注意標記的大小，顏色和透明度。

```python
n = 1024 
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)

plt.scatter(X,Y,c='r',s=20,alpha=0.2)

plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)

plt.show()
```
